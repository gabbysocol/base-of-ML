# Start

```python
class Hierarchy:
    
    def __init__(self, n):
        self.n = n
        
    
    def get_settings(self):
        Z = linkage(self.df, 'single')
        fig = plt.figure(figsize=(25, 20))
        plt.title("Distances in Dendograms", fontsize=22)

        dend = dendrogram(Z, color_threshold=200)
        plt.xticks(fontsize=12)
        plt.show()
#         print(self.Z)
    
        
    def get_settings1(self):
        Z = linkage(self.df, 'complete')
        fig = plt.figure(figsize=(25, 20))
        plt.title("Distances in Dendograms", fontsize=22)

        dend = dendrogram(Z, color_threshold=200)
        plt.xticks(fontsize=12)
        plt.show()
#         print(self.Z)
    
    
    def load_data(self):
        self.df = pd.read_csv('test2.csv', sep=';')
        print(self.df)
        
    
    def get_grafic(self):
        fig = plt.figure(figsize=(25, 20))
        plt.title("Distances in Dendograms", fontsize=22)

        dend = dendrogram(self.Z, color_threshold=200)
        plt.xticks(fontsize=12)
        plt.show()
```
